```python
import pandas as pd

data = {
    "gene": ["TP53", "EGFR", "BRCA1", "BRCA2"],
    "expression": [10.5, 8.2, 5.4, 6.7]
}

df = pd.DataFrame(data)

print("DataFrame:")
print(df)

print("\nExpression values:")
print(df["expression"])

print("\nSummary statistics:")
print(df["expression"].describe())
```

```python
import matplotlib.pyplot as plt

plt.bar(df["gene"], df["expression"], color="skyblue")
plt.xlabel("Gene")
plt.ylabel("Expression Level")
plt.title("Gene Expression Example")
plt.show()
```

```python
import pandas as pd

data = {
    "gene": ["TP53", "EGFR", "BRCA1", "BRCA2"],
    "expression": [10.5, 8.2, 5.4, 6.7]
}

df = pd.DataFrame(data)

df.to_csv("data/sample.csv", index=False)

print("CSV 파일 저장 완료!")
```

```python
df_loaded = pd.read_csv("data/sample.csv")
print(df_loaded)
```

```python
print(df_loaded.describe())
```

```python
import matplotlib.pyplot as plt

plt.boxplot(df_loaded["expression"], labels=["expression"])
plt.title("Expression Boxplot")
plt.ylabel("Expression Level")
plt.show()

plt.hist(df_loaded["expression"], bins=5, color="skyblue", edgecolor="black")
plt.title("Expression Histogram")
plt.xlabel("Expression Level")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.bar(df_loaded["gene"], df_loaded["expression"], color="skyblue")
plt.xlabel("Gene")
plt.ylabel("Expression Level")
plt.title("Gene Expression Example")

plt.ylim(0, 12)

plt.show()
```
